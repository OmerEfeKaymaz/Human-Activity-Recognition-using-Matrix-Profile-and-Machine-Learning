```python
import os
import math

def compute_matrix_profile(file_path, m=50):
    time_series = []

    with open(file_path, 'r') as file:
        for line in file:
            values = line.split(',')
            torso_x = float(values[0])
            time_series.append(torso_x)

    n = len(time_series)
    distance_matrix = []

    for i in range(n - m + 1):
        r_subseq = time_series[i:i + m]
        distance_profile = []

        for j in range(n - m + 1):
            s_subseq = time_series[j:j + m]
            D = sum((r_subseq[k] - s_subseq[k]) ** 2 for k in range(m))
            euclidean_distance = math.sqrt(D)
            distance_profile.append(euclidean_distance)

        distance_matrix.append(distance_profile)

    matrix_profile = [
        min([dist for j, dist in enumerate(row) if j != i])
        for i, row in enumerate(distance_matrix)
    ]

    return matrix_profile

feature_matrix = []
labels = []

base_path = 'daily_sports_activities/data'
for activity in sorted(os.listdir(base_path)):
    activity_path = os.path.join(base_path, activity)
    for subject in ['p1']:
        subject_path = os.path.join(activity_path, subject)
        for segment in sorted(os.listdir(subject_path)):
            file_path = os.path.join(subject_path, segment)

            matrix_profile = compute_matrix_profile(file_path)

            matrix_profile_rounded = [round(num, 2) for num in matrix_profile]

            feature_matrix.append(matrix_profile_rounded)
            labels.append(activity)

for i in range(min(len(feature_matrix), len(labels))):
    feature_vector = feature_matrix[i]
    label = labels[i]
    print(f"Feature Vector {i+1} (Activity: {label}): {feature_vector}")
```

```python
import os
import math
import pandas as pd
import numpy as np

def compute_matrix_profile(file_path, m=50):
    time_series = []

    with open(file_path, 'r') as file:
        for line in file:
            values = line.split(',')
            torso_x = float(values[0])
            time_series.append(torso_x)

    n = len(time_series)
    distance_matrix = []

    for i in range(n - m + 1):
        r_subseq = time_series[i:i + m]
        distance_profile = []

        for j in range(n - m + 1):
            s_subseq = time_series[j:j + m]
            D = sum((r_subseq[k] - s_subseq[k]) ** 2 for k in range(m))
            euclidean_distance = math.sqrt(D)
            distance_profile.append(euclidean_distance)

        distance_matrix.append(distance_profile)

    matrix_profile = [
        min([dist for j, dist in enumerate(row) if j != i])
        for i, row in enumerate(distance_matrix)
    ]

    return matrix_profile

feature_matrix = []
labels = []

base_path = 'daily_sports_activities/data'
for activity in sorted(os.listdir(base_path)):
    activity_path = os.path.join(base_path, activity)
    for subject in ['p1']:
        subject_path = os.path.join(activity_path, subject)
        for segment in sorted(os.listdir(subject_path)):
            file_path = os.path.join(subject_path, segment)

            matrix_profile = compute_matrix_profile(file_path)
            matrix_profile_rounded = [round(num, 2) for num in matrix_profile]

            feature_matrix.append(matrix_profile_rounded)
            labels.append(activity)

matrix_profiles_2d = np.array(feature_matrix)

matrix_profiles_2d = np.reshape(matrix_profiles_2d, (1140, 76))

# DataFrame and CSV
data = pd.DataFrame(matrix_profiles_2d)
data['Activity'] = labels
data.to_csv("activity_data.csv", index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

data = pd.read_csv("activity_data.csv")

# Split test and train
train, test = train_test_split(data, test_size=0.2, random_state=42)

train_y = train['Activity']
train_x = train.drop(['Activity'], axis=1)

test_y = test['Activity']
test_x = test.drop(['Activity'], axis=1)

scaler = StandardScaler()
train_x = scaler.fit_transform(train_x)
test_x = scaler.transform(test_x)

train_scaled = pd.DataFrame(train_x, columns=train.columns[:-1])
train_scaled['Activity'] = train_y.values
train_scaled.to_csv("train.csv", index=False)

test_scaled = pd.DataFrame(test_x, columns=test.columns[:-1])
test_scaled['Activity'] = test_y.values
test_scaled.to_csv("test.csv", index=False)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

train_y = train['Activity']
train_x = train.drop(['Activity'], axis=1)

test_y = test['Activity']
test_x = test.drop(['Activity'], axis=1)

# Hyperparameter grid
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.001, 0.01, 0.1, 1],
    'kernel': ['rbf'],
    'degree': [3],
    'coef0': [0.0],
    'shrinking': [True, False],
    'probability': [False],
    'tol': [0.001],
    'cache_size': [200],
    'class_weight': ['balanced', None],
    'verbose': [False],
    'max_iter': [-1],
    'random_state': [42]
}

# Grid Search with SVC model
svc = SVC(random_state=42)

grid_search = GridSearchCV(svc, param_grid, cv=3, scoring='accuracy')
grid_search.fit(train_x, train_y)

print("Parameters:", grid_search.best_params_)

best_model = grid_search.best_estimator_
predicted = best_model.predict(test_x)

print(f"Accuracy: {accuracy_score(test_y, predicted) * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(test_y, predicted))
```
